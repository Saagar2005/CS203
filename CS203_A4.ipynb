```python
import numpy as np
from scipy.special import erfinv, erf
import matplotlib.pyplot as plt
import seaborn as sns

mu = float(input("Enter the mean of the normal distribution you want to sample from:"))
stdDev = float(input("Enter the standard deviation of the normal distribution you want to sample from:"))
print("Let [a,b] be the range you want to sample from")
a = float(input("Enter the value of a:"))
b = float(input("Enter the value of b:"))

samples_from_uniform = np.random.uniform(size=10000)

required_samples = np.sqrt(2) * erfinv(2*samples_from_uniform - 1)

samples_from_normal = required_samples * stdDev + mu

samples_final = samples_from_normal[((samples_from_normal >= a) & (samples_from_normal <= b))]

sns.histplot(samples_final, bins = 200, kde=True)
plt.xlabel("Sampled value")
plt.ylabel("Probability (Frequency of occurence of this sample/Total number of samples)")
plt.title("Histogram of Samples from Normal distribution")
plt.show()
```
